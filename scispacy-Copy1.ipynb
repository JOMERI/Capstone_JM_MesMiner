```python
def get_entities(document): #This function gets entites from a str document and an nlp en_core_sci_md

   
    doc = nlp(document)
    ent_lst=[]
    for ent in doc.ents:
        #print(ent.text)
        ent_lst.append(ent)         
    return  ent_lst
```

```python
bionlp = spacy.load("en_ner_bionlp13cg_md")
```

```python
def get_genes(document): #This function gets genes, drugs and organism from a str document and an nlp en_ner_bionlp13cg_md

   
    doc = bionlp(document)
    gene_lst = []
    drug_lst = []
    org_lst = []
    
    for ent in doc.ents:
        if ent.label_ =='GENE_OR_GENE_PRODUCT':
            gene_lst.append(ent)
        elif ent.label_ =='SIMPLE_CHEMICAL':
            drug_lst.append(ent)
        elif ent.label_ =='CELL' or ent.label_ =='ORGANISM':
            org_lst.append(ent)
    return  gene_lst,drug_lst,org_lst 

#returns list of lists: 
#[0] ent
#[1] genes
#[2] drugs
#[3] organism/cell for searching propuses
```

```python
def get_genes(document): #This function gets genes, drugs and organism from a str document and an nlp en_ner_bionlp13cg_md

   
    doc = bionlp(document)
    gene_lst = {}
    drug_lst = {}
    org_lst = {}
    
    for ent in doc.ents:
        if ent.label_ =='GENE_OR_GENE_PRODUCT':
            gene_lst[ent]=ent.text
        elif ent.label_ =='SIMPLE_CHEMICAL':
            drug_lst[ent]=ent.text
        elif ent.label_ =='CELL' or ent.label_ =='ORGANISM':
            org_lst[ent]=ent.text
    return  gene_lst,drug_lst,org_lst 

#returns list of lists: 
#[0] ent
#[1] genes
#[2] drugs
#[3] organism/cell for searching propuses
```

```python
def get_entiteis_cmb (srt):
    all_lst = []
    ent = get_entities(srt)
    bio = get_genes(srt)
    all_lst.append(ent)
    for i in range(len(bio)):
        all_lst.append(bio[i])
    return all_lst
```

```python
def get_all_scispacy(lst_txt): #lis_of strigns
    list_genes=[]
    for lst in lst_txt:
        a=get_entiteis_cmb(lst)
        list_genes.append(a)  
    return list_genes
```
